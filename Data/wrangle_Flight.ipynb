```python
def wrangle_data(data, interesting_var, rename_col):
#     drop duplicated data
    data.drop_duplicates(inplace = True)
    
#     fill missing value with 0 in columns CarrierDelay, WeatherDelay, NASDelay, SecurityDelay, and LateAircraftDelay
    delay_col = ['CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']
    data[delay_col] = data[delay_col].fillna(0, axis = 1)
    
#     I won't explore on diverted data; hence, I drop them out
    data.query('Diverted == 0', inplace = True)
    data.drop(['Diverted'], axis = 1, inplace = True)
    data.reset_index(drop = True, inplace = True)
    
#     filter interested features
    data = data.filter(items = interesting_var)
    
#     creat new feature "arrival_status"
    conditions_status = [
        (data['Cancelled'] == 1),
        (data['ArrDelay'] >= 15)]
    choices_status = ['Cancelled','Delay']
    data['arrival_status'] = np.select(conditions_status, choices_status, default = 'Normal')
    data.drop(['Cancelled'], axis = 1, inplace = True)
        
#     rename CancellationCode. A:Carrier, B:Weather, C:NAS, D:Security
    conditions_cancel = [
        (data['CancellationCode'] == 'A'),
        (data['CancellationCode'] == 'B'),
        (data['CancellationCode'] == 'C'),
        (data['CancellationCode'] == 'D')
    ]
    choices_cancel = ['Carrier', 'Weather', 'NAS', 'Security']
    data['CancellationCode'] = np.select(conditions_cancel, choices_cancel, default = 'None')
    
    data.columns = rename_col
    
    return data
```

```python
# Features I choose
interesting_var = ['Year', 'Month', 'DayOfWeek', 'UniqueCarrier', 
                   'ArrDelay', 'Dest', 'Distance', 'Cancelled',
                   'CancellationCode', 'CarrierDelay', 'WeatherDelay', 
                   'NASDelay', 'SecurityDelay', 'LateAircraftDelay']
```

```python
rename_col = ['year', 'month', 'day_of_week', 'carrier', 
            'arrdelay_time', 'destination', 'distance',
            'cancellation_code', 'carrier_delay', 'weather_delay', 
            'nas_delay', 'security_delay', 'late_aircraft_delay', 'arrival_status']
```

```python
df_2004 = wrangle_data(df_2004, interesting_var, rename_col)
df_2005 = wrangle_data(df_2005, interesting_var, rename_col)
df_2006 = wrangle_data(df_2006, interesting_var, rename_col)
df_2007 = wrangle_data(df_2007, interesting_var, rename_col)
df_2008 = wrangle_data(df_2008, interesting_var, rename_col)
```
